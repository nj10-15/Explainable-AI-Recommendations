```python
# =========================================================
# EXPLAINABLE AI MOVIE RECOMMENDER - CORRECTED VERSION
# Research Question: Do explanations increase trust & influence decisions?
# =========================================================

# Install required packages
!pip -q install pandas numpy scipy matplotlib seaborn kagglehub

import os
import re
import time
import random
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import kagglehub
from scipy import stats

# Configuration
warnings.filterwarnings("ignore")
random.seed(42)
np.random.seed(42)

print("✓ Libraries loaded successfully")
```

```python
# =========================================================
# LOAD TMDB DATASET
# =========================================================
print("\n📥 Downloading TMDB dataset...")
path = kagglehub.dataset_download("asaniczka/tmdb-movies-dataset-2023-930k-movies")

# Find the CSV file in downloaded folder
tmdb_file = None
for root, _, files in os.walk(path):
    for f in files:
        if f.lower().endswith(".csv"):
            tmdb_file = os.path.join(root, f)
            break
    if tmdb_file:
        break

if tmdb_file is None:
    raise FileNotFoundError("❌ No CSV file found in TMDB dataset")

# Load TMDB data
tmdb_df = pd.read_csv(tmdb_file)
print(f"✓ TMDB loaded: {tmdb_df.shape[0]:,} movies")

# =========================================================
# LOAD MOVIELENS DATASET
# =========================================================
print("\n📥 Downloading MovieLens dataset...")

import urllib.request
import zipfile

# Download MovieLens
movielens_url = "https://files.grouplens.org/datasets/movielens/ml-latest-small.zip"
movielens_zip = "ml-latest-small.zip"

if not os.path.exists("ml-latest-small"):
    print("Downloading MovieLens...")
    urllib.request.urlretrieve(movielens_url, movielens_zip)
    
    print("Extracting...")
    with zipfile.ZipFile(movielens_zip, 'r') as zip_ref:
        zip_ref.extractall('.')
    
    print("✓ MovieLens downloaded and extracted")

# Load MovieLens files
movies_df = pd.read_csv("ml-latest-small/movies.csv")
ratings_df = pd.read_csv("ml-latest-small/ratings.csv")

print(f"✓ MovieLens loaded: {movies_df.shape[0]:,} movies, {ratings_df.shape[0]:,} ratings")
```

```python
# =========================================================
# HELPER FUNCTIONS
# =========================================================

def normalize_title(text):
    """Clean movie titles for matching (remove year, special chars)"""
    text = str(text).lower()
    text = re.sub(r"\(\d{4}\)", "", text)  # Remove (1999)
    text = re.sub(r"[^a-z0-9\s]", " ", text)  # Keep only letters/numbers
    text = re.sub(r"\s+", " ", text).strip()
    return text

def clean_text(text):
    """Clean text for keyword matching"""
    text = str(text).lower()
    text = re.sub(r"[^a-z0-9\s\-]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

def pick_column(df, candidates, required=True):
    """Find column name from list of possible names (handles dataset variations)"""
    lower_map = {c.lower(): c for c in df.columns}
    for candidate in candidates:
        if candidate.lower() in lower_map:
            return lower_map[candidate.lower()]
    if required:
        raise ValueError(f"Could not find columns: {candidates}")
    return None

print("✓ Helper functions defined")
```

```python
# =========================================================
# PREPROCESS MOVIELENS
# =========================================================
print("\n🔧 Processing MovieLens data...")

# Calculate statistics for each movie
movie_stats = (
    ratings_df.groupby("movieId")
    .agg(
        avg_rating=("rating", "mean"),
        num_ratings=("rating", "count")
    )
    .reset_index()
)

# Merge stats into movies
movies_df = movies_df.merge(movie_stats, on="movieId", how="left")
movies_df["avg_rating"] = movies_df["avg_rating"].fillna(0.0)
movies_df["num_ratings"] = movies_df["num_ratings"].fillna(0).astype(int)

# Extract year from title: "Inception (2010)" → 2010
movies_df["year"] = movies_df["title"].str.extract(r"\((\d{4})\)")
movies_df["year"] = pd.to_numeric(movies_df["year"], errors="coerce")

# Create clean title for matching
movies_df["title_clean"] = movies_df["title"].apply(normalize_title)

# Filter: only keep movies with at least 20 ratings (more reliable)
movies_df = movies_df[movies_df["num_ratings"] >= 20].copy().reset_index(drop=True)

print(f"✓ Filtered to {len(movies_df):,} movies with 20+ ratings")
```

```python
# =========================================================
# PREPARE TMDB DATA
# =========================================================
print("\n🔧 Processing TMDB data...")

# Find column names (different datasets use different names)
title_col = pick_column(tmdb_df, ["title", "movie_title"])
overview_col = pick_column(tmdb_df, ["overview", "description"], required=False)
genre_col = pick_column(tmdb_df, ["genres", "genre"], required=False)
rating_col = pick_column(tmdb_df, ["vote_average", "rating"], required=False)
count_col = pick_column(tmdb_df, ["vote_count", "rating_count", "votes"], required=False)
date_col = pick_column(tmdb_df, ["release_date", "release_year", "year"], required=False)
popularity_col = pick_column(tmdb_df, ["popularity"], required=False)

# Create standardized columns
tmdb_df = tmdb_df.copy()
tmdb_df["title_clean"] = tmdb_df[title_col].apply(normalize_title)

# Overview (movie description)
if overview_col is not None:
    tmdb_df["overview_clean"] = tmdb_df[overview_col].fillna("").astype(str)
else:
    tmdb_df["overview_clean"] = ""

# Genres
if genre_col is not None:
    tmdb_df["tmdb_genres"] = tmdb_df[genre_col].fillna("").astype(str)
else:
    tmdb_df["tmdb_genres"] = ""

# TMDB Rating (0-10 scale)
if rating_col is not None:
    tmdb_df["tmdb_rating"] = pd.to_numeric(tmdb_df[rating_col], errors="coerce").fillna(0.0)
else:
    tmdb_df["tmdb_rating"] = 0.0

# Vote count (how many people rated)
if count_col is not None:
    tmdb_df["tmdb_votes"] = pd.to_numeric(tmdb_df[count_col], errors="coerce").fillna(0.0)
else:
    tmdb_df["tmdb_votes"] = 0.0

# Popularity score
if popularity_col is not None:
    tmdb_df["tmdb_popularity"] = pd.to_numeric(tmdb_df[popularity_col], errors="coerce").fillna(0.0)
else:
    tmdb_df["tmdb_popularity"] = 0.0

# Release year
if date_col is not None:
    tmdb_df["tmdb_year"] = pd.to_datetime(tmdb_df[date_col], errors="coerce").dt.year
    if tmdb_df["tmdb_year"].isna().all():
        tmdb_df["tmdb_year"] = pd.to_numeric(tmdb_df[date_col], errors="coerce")
else:
    tmdb_df["tmdb_year"] = np.nan

print(f"✓ TMDB processed: {len(tmdb_df):,} movies with metadata")
```

```python
# =========================================================
# MERGE TMDB INTO MOVIELENS
# =========================================================
print("\n🔗 Merging datasets...")

# Remove duplicate TMDB entries (keep best version of each movie)
tmdb_df["merge_score"] = (
    tmdb_df["tmdb_rating"] * 0.45 +
    np.log1p(tmdb_df["tmdb_votes"]) * 0.35 +
    np.log1p(tmdb_df["tmdb_popularity"]) * 0.20
)

tmdb_best = (
    tmdb_df.sort_values("merge_score", ascending=False)
    .drop_duplicates(subset=["title_clean"])
    .copy()
)

# Merge TMDB data into MovieLens
movies_df = movies_df.merge(
    tmdb_best[[
        "title_clean",
        "tmdb_rating",
        "tmdb_votes",
        "tmdb_popularity",
        "tmdb_year",
        "tmdb_genres",
        "overview_clean"
    ]],
    on="title_clean",
    how="left"
)

# Fill missing values
movies_df["tmdb_rating"] = movies_df["tmdb_rating"].fillna(0.0)
movies_df["tmdb_votes"] = movies_df["tmdb_votes"].fillna(0.0)
movies_df["tmdb_popularity"] = movies_df["tmdb_popularity"].fillna(0.0)
movies_df["tmdb_year"] = movies_df["tmdb_year"].fillna(movies_df["year"])
movies_df["tmdb_genres"] = movies_df["tmdb_genres"].fillna("")
movies_df["overview_clean"] = movies_df["overview_clean"].fillna("")

matched_count = int((movies_df["tmdb_votes"] > 0).sum())
print(f"✓ Merged! {matched_count:,} movies now have TMDB metadata")
print(f"✓ Final dataset: {len(movies_df):,} movies ready for recommendations")
```

```python
# =========================================================
# TRUST QUESTIONS (Two versions for different conditions)
# =========================================================

# Questions for NO EXPLANATION condition
TRUST_QUESTIONS_NO_EXPLANATION = [
    "I trust this recommendation",
    "The system seems reliable",
    "This seems like a good match for me",
    "I would use this system again",
    "I feel confident about this recommendation"
]

# Questions for WITH EXPLANATION condition
TRUST_QUESTIONS_WITH_EXPLANATION = [
    "I trust this recommendation",
    "The system seems reliable",
    "I understand why this recommendation was made",
    "The explanation is helpful",
    "I would use this system again"
]

# =========================================================
# USER INPUT HELPERS
# =========================================================

def safe_int_input(prompt, min_val=None, max_val=None, allow_blank=False):
    """Get integer input with validation"""
    while True:
        raw = input(prompt).strip()
        
        if allow_blank and raw == "":
            return None
        
        try:
            value = int(raw)
            if min_val is not None and value < min_val:
                print(f"⚠️ Please enter a value >= {min_val}")
                continue
            if max_val is not None and value > max_val:
                print(f"⚠️ Please enter a value <= {max_val}")
                continue
            return value
        except ValueError:
            print("⚠️ Please enter a valid number")

def safe_choice_input(prompt, valid_choices):
    """Get choice input with validation"""
    valid_choices = [str(v).lower() for v in valid_choices]
    while True:
        value = input(prompt).strip().lower()
        if value in valid_choices:
            return value
        print(f"⚠️ Please choose one of: {', '.join(valid_choices)}")

print("✓ Trust questions (2 versions) and input helpers defined")
```

```python
# =========================================================
# COLLECT USER PREFERENCES (5 QUESTIONS)
# =========================================================

def collect_user_preferences():
    """Ask user about their movie preferences"""
    print("\n" + "="*50)
    print("📋 TELL US YOUR MOVIE PREFERENCES")
    print("="*50)
    
    favorite_genre = input(
        "\n1️⃣ Favorite genre (Action, Comedy, Drama, Sci-Fi, Romance, Thriller, etc.): "
    ).strip()
    
    favorite_movie = input(
        "2️⃣ One movie you really like: "
    ).strip()
    
    preferred_year = safe_int_input(
        "3️⃣ Preferred release year (optional, press Enter to skip): ",
        min_val=1900,
        max_val=2030,
        allow_blank=True
    )
    
    mood = input(
        "4️⃣ Current mood (fun, emotional, dark, thrilling, mind-bending, inspiring, etc.): "
    ).strip()
    
    liked_aspects = input(
        "5️⃣ What matters most? (story, acting, visuals, suspense, humor, action, emotion, etc.): "
    ).strip()
    
    return {
        "favorite_genre": favorite_genre,
        "favorite_movie": favorite_movie,
        "preferred_year": preferred_year,
        "mood": mood,
        "liked_aspects": liked_aspects
    }

print("✓ Preference collection function defined")
```

```python
# =========================================================
# FIND USER'S FAVORITE MOVIE IN DATASET
# =========================================================

def find_movie_match(movie_name, movies_data):
    """Find a movie in dataset by name (fuzzy matching)"""
    if not movie_name:
        return None
    
    # Clean the search query
    query = normalize_title(movie_name)
    
    # Search for matches
    matches = movies_data[
        movies_data["title_clean"].str.contains(query, case=False, na=False)
    ].copy()
    
    if matches.empty:
        return None
    
    # Return the best match (highest rated + most ratings)
    matches = matches.sort_values(
        by=["avg_rating", "num_ratings"],
        ascending=[False, False]
    )
    return matches.iloc[0]

print("✓ Movie matching function defined")
```

```python
# =========================================================
# PERSONALIZED RECOMMENDATION ENGINE
# =========================================================

def generate_personalized_recommendations(preferences, movies_data, top_n=6):
    """Generate personalized movie recommendations based on user preferences"""
    
    df = movies_data.copy()
    df["score"] = 0.0
    
    favorite_genre = preferences["favorite_genre"]
    favorite_movie = preferences["favorite_movie"]
    preferred_year = preferences["preferred_year"]
    
    # SCORING COMPONENT 1: Genre Match (30% weight)
    if favorite_genre:
        df["genre_match_ml"] = df["genres"].str.contains(
            favorite_genre, case=False, na=False
        ).astype(int)
        
        df["genre_match_tmdb"] = df["tmdb_genres"].str.contains(
            favorite_genre, case=False, na=False
        ).astype(int)
        
        df["genre_match"] = np.maximum(df["genre_match_ml"], df["genre_match_tmdb"])
        df["score"] += df["genre_match"] * 0.30
    else:
        df["genre_match"] = 0
    
    # SCORING COMPONENT 2: Similarity to Favorite Movie (18% weight)
    favorite_movie_row = find_movie_match(favorite_movie, movies_data)
    if favorite_movie_row is not None:
        fav_genres = set(str(favorite_movie_row["genres"]).split("|"))
        
        def overlap_count(genres):
            current = set(str(genres).split("|"))
            return len(current.intersection(fav_genres))
        
        df["fav_similarity"] = df["genres"].apply(overlap_count)
        df["score"] += df["fav_similarity"] * 0.18
        
        # Remove the favorite movie itself from recommendations
        df = df[df["movieId"] != favorite_movie_row["movieId"]]
    else:
        df["fav_similarity"] = 0
        favorite_movie_row = None
    
    # SCORING COMPONENT 3: Year Preference (8% weight)
    if preferred_year is not None:
        year_ref = df["tmdb_year"].fillna(df["year"])
        df["year_distance"] = (year_ref - preferred_year).abs()
        df["year_distance"] = df["year_distance"].fillna(100)
        df["year_match"] = np.maximum(0, 1 - (df["year_distance"] / 20))
        df["score"] += df["year_match"] * 0.08
    else:
        df["year_match"] = 0
    
    # SCORING COMPONENT 4: MovieLens Rating (14% weight)
    df["rating_norm"] = df["avg_rating"] / 5.0
    df["score"] += df["rating_norm"] * 0.14
    
    # SCORING COMPONENT 5: TMDB Rating (14% weight)
    df["tmdb_rating_norm"] = df["tmdb_rating"] / 10.0
    df["score"] += df["tmdb_rating_norm"] * 0.14
    
    # SCORING COMPONENT 6: MovieLens Popularity (6% weight)
    max_pop_ml = max(df["num_ratings"].max(), 1)
    df["popularity_norm_ml"] = df["num_ratings"] / max_pop_ml
    df["score"] += df["popularity_norm_ml"] * 0.06
    
    # SCORING COMPONENT 7: TMDB Popularity (5% weight)
    max_pop_tmdb = max(df["tmdb_popularity"].max(), 1)
    df["popularity_norm_tmdb"] = df["tmdb_popularity"] / max_pop_tmdb
    df["score"] += df["popularity_norm_tmdb"] * 0.05
    
    # SCORING COMPONENT 8: TMDB Votes (3% weight)
    max_votes = max(df["tmdb_votes"].max(), 1)
    df["vote_norm"] = df["tmdb_votes"] / max_votes
    df["score"] += df["vote_norm"] * 0.03
    
    # Sort by score and return top N
    df = df.sort_values("score", ascending=False)
    final_recs = df.head(top_n).reset_index(drop=True)
    
    return final_recs, favorite_movie_row

print("✓ Recommendation engine defined")
print("   Scoring weights: Genre(30%) + Similarity(18%) + Rating(28%) + Year(8%) + Popularity(14%)")
```

```python
# =========================================================
# EXPLANATION GENERATORS (3 DISTINCT TYPES - HIGHLY DIFFERENTIATED)
# Target length: ~700 characters (6-8 detailed points)
# NOTE: Do NOT repeat ratings/votes already shown in movie display
# =========================================================

def generate_rating_explanation(movie_row):
    """
    RATING-BASED EXPLANATION
    Focus: WHY it's highly rated - awards, critical acclaim, technical excellence
    Does NOT mention: TMDB/MovieLens scores (already displayed above)
    Uses: Awards, critics, legacy, technical aspects, director/cast
    Target: ~700 chars
    """
    title = movie_row["title"]
    tmdb_rating = float(movie_row.get("tmdb_rating", 0.0))
    tmdb_votes = int(movie_row.get("tmdb_votes", 0))
    year = movie_row.get("tmdb_year", movie_row.get("year", ""))
    overview = str(movie_row.get("overview_clean", ""))
    genres = str(movie_row.get("genres", ""))
    
    explanation_parts = []
    
    # WHY it's critically acclaimed (not the score itself)
    if tmdb_rating >= 8.5:
        explanation_parts.append(
            "Widely considered a masterpiece of cinema - critics praised its innovative storytelling, "
            "exceptional performances, and groundbreaking technical achievements that set new standards for the industry"
        )
    elif tmdb_rating >= 8.0:
        explanation_parts.append(
            "Critically acclaimed for its outstanding direction, compelling narrative structure, "
            "and memorable performances that have influenced countless filmmakers"
        )
    elif tmdb_rating >= 7.5:
        explanation_parts.append(
            "Highly praised by critics for its strong script, skilled direction, and impressive production values "
            "that elevate it above standard genre fare"
        )
    elif tmdb_rating >= 7.0:
        explanation_parts.append(
            "Well-received by critics who highlighted its solid craftsmanship, engaging storytelling, "
            "and professional execution across all technical departments"
        )
    else:
        explanation_parts.append(
            "Recognized for its competent filmmaking and genuine effort to deliver quality entertainment"
        )
    
    # Awards and recognition (simulated based on rating + votes)
    if tmdb_rating >= 8.0 and tmdb_votes >= 100000:
        explanation_parts.append(
            "Multiple Academy Award nominations including Best Picture, with wins for technical categories. "
            "Featured prominently in AFI's Top 100 Films and Sight & Sound's critical polls"
        )
    elif tmdb_rating >= 8.0 and tmdb_votes >= 50000:
        explanation_parts.append(
            "Recipient of major industry awards including Golden Globe recognition. "
            "Consistently appears in 'Best Films of All Time' lists by major publications"
        )
    elif tmdb_rating >= 7.5 and tmdb_votes >= 20000:
        explanation_parts.append(
            "Earned critical award nominations and won recognition at major film festivals. "
            "Frequently cited as essential viewing in its genre"
        )
    elif tmdb_rating >= 7.0:
        explanation_parts.append(
            "Received positive recognition from film critics' associations and genre-specific awards"
        )
    
    # Legacy and cultural impact based on age + rating
    if year and not pd.isna(year):
        year_int = int(year)
        current_year = 2026
        age = current_year - year_int
        
        if age >= 30 and tmdb_rating >= 7.5:
            explanation_parts.append(
                f"A {year_int} classic that has stood the test of time - its influence can be seen in modern cinema, "
                f"and film schools continue to study its techniques and narrative innovations"
            )
        elif age >= 20 and tmdb_rating >= 7.5:
            explanation_parts.append(
                f"Released in {year_int}, this film defined its era and continues to be referenced in contemporary "
                f"filmmaking. Its impact on the genre remains significant decades later"
            )
        elif age >= 10 and tmdb_rating >= 7.5:
            explanation_parts.append(
                f"This {year_int} release quickly became recognized as a modern classic, with critics noting its "
                f"sophisticated approach and technical excellence that will age well"
            )
        elif age >= 5 and tmdb_rating >= 7.0:
            explanation_parts.append(
                f"Recent {year_int} release that earned immediate critical recognition for its fresh perspective "
                f"and quality craftsmanship"
            )
    
    # Technical/artistic excellence from genres
    if "Drama" in genres:
        explanation_parts.append(
            "The performances deliver emotional depth and authenticity, with actors bringing nuanced "
            "character work that critics highlighted as career-best"
        )
    elif any(g in genres for g in ["Sci-Fi", "Fantasy", "Action"]):
        explanation_parts.append(
            "Showcases impressive visual effects and cinematography that critics described as 'groundbreaking' "
            "and 'visually stunning', with technical execution that pushed boundaries"
        )
    elif "Thriller" in genres or "Mystery" in genres:
        explanation_parts.append(
            "Expertly crafted suspense and narrative pacing that critics praised for maintaining tension "
            "while delivering intelligent storytelling"
        )
    
    # Plot with critical perspective
    if overview and len(overview) > 50:
        first_sentence = overview.split('.')[0][:140]
        if len(first_sentence) > 30:
            explanation_parts.append(
                f"The story - {first_sentence}... - has been analyzed for its thematic depth and "
                f"sophisticated narrative structure"
            )
    
    # Join parts for ~700 character target
    explanation = f"🏆 {title}: " + " ".join(explanation_parts[:4]) + "."
    
    return "rating", explanation


def generate_similarity_explanation(movie_row, preferences, favorite_movie_row):
    """
    SIMILARITY-BASED EXPLANATION
    Focus: Personal connections to user's taste, favorite movie, preferences
    Does NOT mention: TMDB/MovieLens scores (already displayed above)
    Uses: Genre DNA, thematic parallels, mood alignment, favorite movie comparison
    Target: ~700 chars
    """
    title = movie_row["title"]
    genres = str(movie_row["genres"])
    year = movie_row.get("tmdb_year", movie_row.get("year", ""))
    overview = str(movie_row.get("overview_clean", ""))
    
    favorite_movie = preferences.get("favorite_movie", "")
    favorite_genre = preferences.get("favorite_genre", "")
    liked_aspects = preferences.get("liked_aspects", "")
    mood = preferences.get("mood", "")
    preferred_year = preferences.get("preferred_year", None)
    
    explanation_parts = []
    
    # Deep connection to favorite movie
    if favorite_movie_row is not None:
        fav_title = favorite_movie_row["title"]
        fav_genres_list = str(favorite_movie_row["genres"]).split("|")
        current_genres_list = genres.split("|")
        
        fav_genres = set(fav_genres_list)
        current_genres = set(current_genres_list)
        shared_genres = fav_genres.intersection(current_genres)
        
        if len(shared_genres) >= 3:
            genre_list = ", ".join(list(shared_genres))
            explanation_parts.append(
                f"Since you loved {fav_title}, this is a perfect match - both films share the {genre_list} DNA, "
                f"exploring similar themes of identity, power, and moral complexity with the same narrative intensity. "
                f"Fans of one typically cite the other as equally compelling"
            )
        elif len(shared_genres) >= 2:
            genre_list = ", ".join(shared_genres)
            explanation_parts.append(
                f"Strong thematic parallel to {fav_title} - both are {genre_list} films that tackle similar "
                f"subject matter with comparable depth and sophistication. If you appreciated the narrative approach "
                f"in one, you'll find the same intelligence here"
            )
        elif len(shared_genres) >= 1:
            genre = list(shared_genres)[0]
            explanation_parts.append(
                f"Connects to {fav_title} through its {genre} foundation - while different in setting, "
                f"both films share a similar storytelling philosophy and emotional core that will resonate with you"
            )
        else:
            explanation_parts.append(
                f"Though different in genre from {fav_title}, this explores complementary themes and will appeal "
                f"to the same sensibilities that drew you to that film - both reward thoughtful viewers"
            )
    elif favorite_movie:
        explanation_parts.append(
            f"Based on what you valued in {favorite_movie}, this film offers a similar viewing experience "
            f"with comparable depth and the kind of storytelling approach you've shown preference for"
        )
    
    # Genre and preference alignment
    if favorite_genre:
        all_genres = genres.replace("|", " and ")
        if favorite_genre.lower() in genres.lower():
            explanation_parts.append(
                f"Perfectly tailored to your {favorite_genre} preference - this is a quintessential example of the genre, "
                f"blending {all_genres} elements in exactly the way {favorite_genre} enthusiasts appreciate most. "
                f"It delivers what you're looking for"
            )
    
    # Mood and tone matching with specifics
    if mood and overview:
        mood_lower = mood.lower()
        
        if "dark" in mood_lower or "intense" in mood_lower:
            explanation_parts.append(
                f"The film's atmosphere matches your {mood} mood perfectly - expect morally complex characters, "
                f"shadowy cinematography, and a narrative that doesn't shy away from uncomfortable truths. "
                f"This isn't feel-good entertainment; it's psychologically engaging"
            )
        elif "fun" in mood_lower or "exciting" in mood_lower:
            explanation_parts.append(
                f"Delivers the {mood} entertainment you're seeking - packed with energy, witty dialogue, "
                f"and genuinely entertaining moments that keep you engaged without demanding heavy emotional investment"
            )
        elif "emotional" in mood_lower:
            explanation_parts.append(
                f"Matches your {mood} mood with genuine emotional resonance - this isn't manipulative melodrama "
                f"but authentic character work that will genuinely move you if you're open to that experience"
            )
        elif "thrilling" in mood_lower:
            explanation_parts.append(
                f"Perfect for your {mood} mood - expertly builds tension and maintains suspense throughout, "
                f"with plot developments that will genuinely surprise and engage you"
            )
    
    # Aspects they value
    if liked_aspects:
        aspects_lower = liked_aspects.lower()
        if "visual" in aspects_lower or "cinematography" in aspects_lower:
            explanation_parts.append(
                f"Since you prioritize {liked_aspects}, you'll appreciate this film's exceptional visual composition - "
                f"every frame is carefully crafted, with cinematography that serves the story while being beautiful in its own right"
            )
        elif "story" in aspects_lower or "plot" in aspects_lower:
            explanation_parts.append(
                f"Given your focus on {liked_aspects}, this delivers exactly what you want - a tightly constructed narrative "
                f"with meaningful character arcs and plot developments that feel earned rather than contrived"
            )
        elif "acting" in aspects_lower or "performance" in aspects_lower:
            explanation_parts.append(
                f"With {liked_aspects} as your priority, you'll value the performances here - actors fully inhabit their "
                f"roles with subtle, layered work that reveals more on repeated viewings"
            )
    
    # Year preference connection
    if preferred_year and year and not pd.isna(year):
        year_int = int(year)
        if abs(year_int - preferred_year) <= 3:
            explanation_parts.append(
                f"Released in {year_int}, this falls right in your preferred era when filmmaking had the style and "
                f"sensibility you clearly appreciate"
            )
        elif abs(year_int - preferred_year) <= 8:
            explanation_parts.append(
                f"From {year_int}, close to your preferred period - carries similar production values and storytelling "
                f"conventions to films from that time"
            )
    
    # Story connection
    if overview and len(overview) > 50:
        hook = overview.split('.')[0][:130]
        if len(hook) > 30:
            explanation_parts.append(
                f"The premise: {hook}... This setup should immediately appeal to you given your stated preferences"
            )
    
    # Join parts for ~700 character target
    explanation = f"❤️ {title}: " + " ".join(explanation_parts[:4]) + "."
    
    return "similarity", explanation


def generate_popularity_explanation(movie_row, preferences):
    """
    POPULARITY-BASED EXPLANATION
    Focus: Social proof, cultural phenomenon, what viewers say, trending status
    Does NOT mention: Raw vote counts or rating numbers (already displayed above)
    Uses: Trending, cultural impact, viewer testimonials, community consensus
    Target: ~700 chars
    """
    title = movie_row["title"]
    tmdb_votes = int(movie_row.get("tmdb_votes", 0))
    tmdb_popularity = float(movie_row.get("tmdb_popularity", 0.0))
    tmdb_rating = float(movie_row.get("tmdb_rating", 0.0))
    ml_count = int(movie_row["num_ratings"])
    year = movie_row.get("tmdb_year", movie_row.get("year", ""))
    genres = str(movie_row["genres"])
    overview = str(movie_row.get("overview_clean", ""))
    
    favorite_genre = preferences.get("favorite_genre", "")
    
    explanation_parts = []
    
    # Trending and cultural moment
    if tmdb_popularity >= 100:
        explanation_parts.append(
            "Currently dominating streaming charts worldwide - this is what everyone is watching and talking about right now. "
            "Social media is flooded with reactions, theories, and recommendations. It's become a genuine cultural moment "
            "where you'll feel left out of conversations if you haven't seen it"
        )
    elif tmdb_popularity >= 50:
        explanation_parts.append(
            "Rapidly climbing popularity charts as word-of-mouth spreads - viewership is surging as people discover it "
            "and immediately tell their friends. This is building into a cultural phenomenon in real-time"
        )
    elif tmdb_popularity >= 20:
        explanation_parts.append(
            "Maintaining strong, consistent popularity across all demographics - this has staying power beyond initial hype. "
            "People keep coming back to it and recommending it months or years after release"
        )
    else:
        explanation_parts.append(
            "Beloved by those who've discovered it - this has a dedicated following who consider it underrated "
            "and passionately recommend it to others"
        )
    
    # Community consensus and viewer behavior
    if favorite_genre and favorite_genre.lower() in genres.lower():
        explanation_parts.append(
            f"Within the {favorite_genre} community, this is considered essential viewing - fan forums, subreddits, "
            f"and discussion groups consistently rank it among the must-watch titles. {favorite_genre} enthusiasts "
            f"cite it as a defining example of what the genre can achieve"
        )
    
    # What viewers are actually saying (simulate testimonials)
    if tmdb_votes >= 100000:
        explanation_parts.append(
            "Viewer consensus is remarkably unified - across hundreds of thousands of reviews, people describe being "
            "'completely absorbed,' 'thinking about it for days after,' and 'immediately rewatching to catch details they missed.' "
            "The recommendation rate among viewers who've seen it is exceptionally high"
        )
    elif tmdb_votes >= 50000:
        explanation_parts.append(
            "The viewing community has spoken with overwhelming positivity - common reactions include 'exceeded expectations,' "
            "'surprised by how good it was,' and 'can't believe more people haven't seen this.' Viewers actively tell others to watch it"
        )
    elif tmdb_votes >= 10000:
        explanation_parts.append(
            "Those who've watched it are enthusiastic advocates - viewer testimonials emphasize being 'genuinely surprised,' "
            "'emotionally invested,' and 'recommending it to everyone.' The fan base may be smaller but is notably passionate"
        )
    
    # Cultural impact and legacy
    if year and not pd.isna(year):
        year_int = int(year)
        current_year = 2026
        age = current_year - year_int
        
        if age >= 20 and tmdb_votes >= 100000:
            explanation_parts.append(
                f"Since its {year_int} release, this has transcended being 'just a movie' to become a cultural reference point - "
                f"quotes from it are part of everyday language, scenes have been parodied countless times, and it appears on "
                f"every 'greatest films' list. New generations continue discovering and embracing it"
            )
        elif age >= 10 and tmdb_votes >= 50000:
            explanation_parts.append(
                f"In the years since {year_int}, this has built a massive following that continues growing - it's become "
                f"the film people reference when recommending others. Its influence on subsequent films is undeniable"
            )
        elif age >= 5 and tmdb_votes >= 20000:
            explanation_parts.append(
                f"This {year_int} release quickly became a modern favorite that people return to repeatedly - "
                f"it's already being called a 'modern classic' and its fan base grows steadily year after year"
            )
        elif age <= 2:
            explanation_parts.append(
                f"As a very recent {year_int} release, it's still building momentum but early viewer response has been "
                f"exceptionally strong - this is on track to become a long-term favorite"
            )
    
    # Social proof from lists and rankings
    if tmdb_votes >= 100000:
        explanation_parts.append(
            "Featured on every major 'must-watch' list - Empire Magazine's Essential Films, IMDb Top 250, "
            "Rotten Tomatoes' Certified Fresh Classics, and countless year-end best-of lists. When critics and "
            "audiences both agree, that consensus is meaningful"
        )
    elif tmdb_votes >= 50000:
        explanation_parts.append(
            "Appears regularly in curated recommendation lists by major publications, streaming platforms, and "
            "film critics. The algorithmic and human-curated recommendations both point to this consistently"
        )
    
    # What viewers love (from overview)
    if overview and len(overview) > 50:
        plot_hook = overview.split('.')[0][:130]
        if len(plot_hook) > 30:
            explanation_parts.append(
                f"What draws viewers in: {plot_hook}... This premise alone has people intrigued, and the execution "
                f"apparently exceeds expectations based on viewer reactions"
            )
    
    # Join parts for ~700 character target
    explanation = f"📈 {title}: " + " ".join(explanation_parts[:3]) + "."
    
    return "popularity", explanation

print("✓ Three HIGHLY DIFFERENTIATED explanation generators defined (~700 chars each):")
print("   🏆 RATING: Awards, critical acclaim, legacy, technical excellence (NO scores)")
print("   ❤️ SIMILARITY: Personal connections, favorite movie parallels, mood/taste match (NO scores)")
print("   📈 POPULARITY: Trending, cultural impact, viewer testimonials, community buzz (NO scores)")
print("   ⚠️ NOTE: Ratings/votes NOT repeated (already shown in movie display)")
```

```python
# =========================================================
# ASSIGN EXPLANATION TYPES TO MOVIES
# =========================================================

def assign_explanation_types(movies_list):
    """
    Randomly assign explanation types to 6 movies:
    - 2 movies get RATING explanations
    - 2 movies get SIMILARITY explanations
    - 2 movies get POPULARITY explanations
    """
    
    # Create list of types (2 of each)
    types = ["rating", "rating", "similarity", "similarity", "popularity", "popularity"]
    
    # Shuffle for randomization
    random.shuffle(types)
    
    # Assign to movies
    assignments = {}
    for i, movie in enumerate(movies_list):
        movie_id = movie["movieId"]
        assignments[movie_id] = types[i]
    
    return assignments

print("✓ Explanation type assignment function defined")
```

```python
# =========================================================
# GENERATE ALL EXPLANATIONS FOR MOVIES
# =========================================================

def generate_all_explanations(movies_df, explanation_assignments, preferences, favorite_movie_row):
    """
    Generate explanations for all movies based on their assigned types
    Returns: dict mapping movieId -> (type, explanation_text)
    """
    
    explanations = {}
    
    for _, movie in movies_df.iterrows():
        movie_id = movie["movieId"]
        exp_type = explanation_assignments.get(movie_id, "rating")
        
        if exp_type == "rating":
            exp_type, exp_text = generate_rating_explanation(movie)
        elif exp_type == "similarity":
            exp_type, exp_text = generate_similarity_explanation(
                movie, preferences, favorite_movie_row
            )
        elif exp_type == "popularity":
            exp_type, exp_text = generate_popularity_explanation(movie, preferences)
        else:
            exp_type, exp_text = "rating", f"Recommended: {movie['title']}"
        
        explanations[movie_id] = (exp_type, exp_text)
    
    return explanations

print("✓ Explanation generation function defined")
```

```python
# =========================================================
# DISPLAY FUNCTIONS
# =========================================================

def display_movie(movie_row, show_explanation, explanation_text, movie_num, total_movies, condition_name):
    """Display a movie recommendation to the user"""
    
    print("\n" + "="*60)
    print(f"📽️  MOVIE {movie_num}/{total_movies} - {condition_name}")
    print("="*60)
    print(f"\n🎬 Title: {movie_row['title']}")
    print(f"🎭 Genres: {movie_row['genres']}")
    print(f"⭐ MovieLens Rating: {movie_row['avg_rating']:.1f}/5 ({int(movie_row['num_ratings'])} ratings)")
    
    if movie_row.get('tmdb_rating', 0) > 0:
        print(f"🌟 TMDB Rating: {float(movie_row['tmdb_rating']):.1f}/10 ({int(movie_row.get('tmdb_votes', 0)):,} votes)")
    
    if show_explanation:
        print(f"\n💡 EXPLANATION:")
        print(f"   {explanation_text}")
    else:
        print(f"\n💡 EXPLANATION: [Hidden for this round]")
    
    print("="*60)

print("✓ Display function defined")
```

```python
# =========================================================
# DATA COLLECTION FUNCTIONS
# =========================================================

def collect_selection_decision():
    """Ask if user would watch this movie"""
    start_time = time.time()
    
    decision = safe_choice_input(
        "\n❓ Would you watch this movie? (y/n): ",
        ["y", "n"]
    )
    
    end_time = time.time()
    decision_time = end_time - start_time
    
    decision_value = 1 if decision == "y" else 0
    
    return decision_value, decision_time


def collect_trust_ratings(explanation_was_shown):
    """
    Collect 5 trust ratings on 1-5 scale
    Uses DIFFERENT questions depending on whether explanation was shown
    
    Args:
        explanation_was_shown: Boolean - True if explanation was displayed
    
    Returns:
        trust_scores: List of 5 integer scores (1-5)
        trust_mean: Float - average of the 5 scores
    """
    
    # Select appropriate question set
    if explanation_was_shown:
        questions = TRUST_QUESTIONS_WITH_EXPLANATION
    else:
        questions = TRUST_QUESTIONS_NO_EXPLANATION
    
    print("\n📊 Please rate the following statements (1-5):")
    print("   1 = Strongly Disagree")
    print("   5 = Strongly Agree")
    print()
    
    trust_scores = []
    
    for i, question in enumerate(questions, 1):
        score = safe_int_input(
            f"   {i}. {question}: ",
            min_val=1,
            max_val=5
        )
        trust_scores.append(score)
    
    trust_mean = np.mean(trust_scores)
    
    return trust_scores, trust_mean

print("✓ Data collection functions defined (with conditional trust questions)")
```

```python
# =========================================================
# RUN ONE CONDITION (6 MOVIES)
# =========================================================

def run_condition_session(
    movies_df,
    explanations_dict,
    show_explanations,
    condition_name
):
    """
    Run one condition session (6 movies with or without explanations)
    
    Args:
        movies_df: DataFrame of 6 movies to show
        explanations_dict: Pre-generated explanations for each movie
        show_explanations: True/False - whether to show explanations
        condition_name: "WITHOUT Explanations" or "WITH Explanations"
    
    Returns:
        List of results (one dict per movie)
    """
    
    results = []
    total_movies = len(movies_df)
    
    print(f"\n{'='*60}")
    print(f"🎯 STARTING: {condition_name}")
    print(f"{'='*60}")
    print(f"You will see {total_movies} movie recommendations.")
    print(f"For each movie, you'll decide if you'd watch it and rate your trust.")
    print()
    input("Press Enter to begin...")
    
    for idx, (_, movie) in enumerate(movies_df.iterrows(), 1):
        movie_id = movie["movieId"]
        exp_type, exp_text = explanations_dict.get(movie_id, ("rating", "Recommended"))
        
        # Display the movie
        display_movie(
            movie_row=movie,
            show_explanation=show_explanations,
            explanation_text=exp_text,
            movie_num=idx,
            total_movies=total_movies,
            condition_name=condition_name
        )
        
        # Collect decision
        decision, decision_time = collect_selection_decision()
        
        # Collect trust ratings (using appropriate question set based on condition)
        trust_scores, trust_mean = collect_trust_ratings(explanation_was_shown=show_explanations)
        
        # Store results
        result = {
            "movieId": movie_id,
            "title": movie["title"],
            "genres": movie["genres"],
            "avg_rating": float(movie["avg_rating"]),
            "num_ratings": int(movie["num_ratings"]),
            "tmdb_rating": float(movie.get("tmdb_rating", 0.0)),
            "tmdb_votes": int(movie.get("tmdb_votes", 0)),
            "explanation_type": exp_type,
            "explanation_shown": show_explanations,
            "explanation_text": exp_text if show_explanations else "",
            "decision": decision,
            "decision_time": decision_time,
            "trust_q1": trust_scores[0],
            "trust_q2": trust_scores[1],
            "trust_q3": trust_scores[2],
            "trust_q4": trust_scores[3],
            "trust_q5": trust_scores[4],
            "trust_mean": trust_mean
        }
        
        results.append(result)
        
        print(f"\n✓ Movie {idx}/{total_movies} complete")
    
    print(f"\n{'='*60}")
    print(f"✅ {condition_name} - COMPLETE!")
    print(f"{'='*60}\n")
    
    return results

print("✓ Condition session function defined (with conditional trust questions)")
```

```python
# =========================================================
# MAIN STUDY FLOW (ONE PARTICIPANT)
# =========================================================

def run_complete_study(participant_id, movies_data):
    """
    Run complete study for one participant
    
    Flow:
    1. Collect preferences
    2. Generate 6 personalized movies
    3. Assign explanation types
    4. Run Condition 1 (with/without explanations based on counterbalancing)
    5. Run Condition 2 (opposite of Condition 1)
    6. Return all data
    """
    
    print("\n" + "🎬"*30)
    print(f"{'='*60}")
    print(f"   EXPLAINABLE AI STUDY - PARTICIPANT {participant_id}")
    print(f"{'='*60}")
    print("🎬"*30)
    
    # Step 1: Collect preferences
    preferences = collect_user_preferences()
    
    # Step 2: Generate personalized recommendations
    print("\n🔍 Finding perfect movies for you...")
    recommended_movies, favorite_movie_row = generate_personalized_recommendations(
        preferences=preferences,
        movies_data=movies_data,
        top_n=6
    )
    
    print(f"✓ Found {len(recommended_movies)} personalized recommendations!")
    
    # Step 3: Assign explanation types (2 rating, 2 similarity, 2 popularity)
    explanation_assignments = assign_explanation_types(
        recommended_movies.to_dict('records')
    )
    
    # Step 4: Generate all explanations
    explanations_dict = generate_all_explanations(
        movies_df=recommended_movies,
        explanation_assignments=explanation_assignments,
        preferences=preferences,
        favorite_movie_row=favorite_movie_row
    )
    
    print("\n✓ Explanations generated:")
    for movie_id, (exp_type, _) in explanations_dict.items():
        movie_title = recommended_movies[recommended_movies['movieId'] == movie_id]['title'].values[0]
        print(f"   • {movie_title}: {exp_type.upper()} explanation")
    
    # Step 5: ALL participants get No Explanation First
    condition_order = "No_Explanation_First"
    first_condition_name = "ROUND 1: WITHOUT Explanations"
    second_condition_name = "ROUND 2: WITH Explanations"
    first_show_exp = False
    second_show_exp = True

    print(f"\n📋 Condition order: {condition_order} (same for all participants)")
    
    # Step 6: Run Condition 1
    condition1_results = run_condition_session(
        movies_df=recommended_movies,
        explanations_dict=explanations_dict,
        show_explanations=first_show_exp,
        condition_name=first_condition_name
    )
    
    # Brief pause between conditions
    print("\n" + "⏸️ "*30)
    print("You've completed the first round!")
    print("Next, you'll see the SAME movies again in a different format.")
    input("Press Enter when ready to continue...")
    print("⏸️ "*30 + "\n")
    
    # Step 7: Run Condition 2 (same movies, opposite condition)
    condition2_results = run_condition_session(
        movies_df=recommended_movies,
        explanations_dict=explanations_dict,
        show_explanations=second_show_exp,
        condition_name=second_condition_name
    )
    
    # Step 8: Compile all data
    participant_data = {
        "participant_id": participant_id,
        "condition_order": condition_order,
        "favorite_genre": preferences["favorite_genre"],
        "favorite_movie": preferences["favorite_movie"],
        "preferred_year": preferences["preferred_year"],
        "mood": preferences["mood"],
        "liked_aspects": preferences["liked_aspects"],
        "condition1_results": condition1_results,
        "condition2_results": condition2_results
    }
    
    print("\n" + "🎉"*30)
    print("✅ STUDY COMPLETE! Thank you for participating!")
    print("🎉"*30 + "\n")
    
    return participant_data

print("✓ Main study flow defined")
```

```python
# =========================================================
# SAVE DATA TO CSV
# =========================================================

def save_participant_data(participant_data, filename="xai_study_results.csv"):
    """
    Convert participant data to flat format and save to CSV
    """
    
    # Extract participant-level info
    pid = participant_data["participant_id"]
    
    # Combine both conditions into one flat structure
    rows = []
    
    # Process condition 1 results
    for movie_result in participant_data["condition1_results"]:
        row = {
            "participant_id": pid,
            "condition_order": participant_data["condition_order"],
            "favorite_genre": participant_data["favorite_genre"],
            "favorite_movie": participant_data["favorite_movie"],
            "preferred_year": participant_data["preferred_year"],
            "mood": participant_data["mood"],
            "liked_aspects": participant_data["liked_aspects"],
            
            "condition": "condition1",
            "explanation_shown": movie_result["explanation_shown"],
            
            "movieId": movie_result["movieId"],
            "title": movie_result["title"],
            "genres": movie_result["genres"],
            "avg_rating": movie_result["avg_rating"],
            "tmdb_rating": movie_result["tmdb_rating"],
            "explanation_type": movie_result["explanation_type"],
            "explanation_text": movie_result["explanation_text"],
            
            "decision": movie_result["decision"],
            "decision_time": movie_result["decision_time"],
            "trust_q1": movie_result["trust_q1"],
            "trust_q2": movie_result["trust_q2"],
            "trust_q3": movie_result["trust_q3"],
            "trust_q4": movie_result["trust_q4"],
            "trust_q5": movie_result["trust_q5"],
            "trust_mean": movie_result["trust_mean"]
        }
        rows.append(row)
    
    # Process condition 2 results
    for movie_result in participant_data["condition2_results"]:
        row = {
            "participant_id": pid,
            "condition_order": participant_data["condition_order"],
            "favorite_genre": participant_data["favorite_genre"],
            "favorite_movie": participant_data["favorite_movie"],
            "preferred_year": participant_data["preferred_year"],
            "mood": participant_data["mood"],
            "liked_aspects": participant_data["liked_aspects"],
            
            "condition": "condition2",
            "explanation_shown": movie_result["explanation_shown"],
            
            "movieId": movie_result["movieId"],
            "title": movie_result["title"],
            "genres": movie_result["genres"],
            "avg_rating": movie_result["avg_rating"],
            "tmdb_rating": movie_result["tmdb_rating"],
            "explanation_type": movie_result["explanation_type"],
            "explanation_text": movie_result["explanation_text"],
            
            "decision": movie_result["decision"],
            "decision_time": movie_result["decision_time"],
            "trust_q1": movie_result["trust_q1"],
            "trust_q2": movie_result["trust_q2"],
            "trust_q3": movie_result["trust_q3"],
            "trust_q4": movie_result["trust_q4"],
            "trust_q5": movie_result["trust_q5"],
            "trust_mean": movie_result["trust_mean"]
        }
        rows.append(row)
    
    # Convert to DataFrame
    df = pd.DataFrame(rows)
    
    # Save to CSV (append if file exists)
    if os.path.exists(filename):
        df.to_csv(filename, mode='a', header=False, index=False)
        print(f"✓ Data appended to {filename}")
    else:
        df.to_csv(filename, index=False)
        print(f"✓ Data saved to {filename}")
    
    return df

print("✓ Data saving function defined")
```

```python
# =========================================================
# RUN THE STUDY
# =========================================================

# Ask how many participants
num_participants = safe_int_input(
    "\n👥 How many participants will complete the study? ",
    min_val=1,
    max_val=50
)

print(f"\n✓ Starting study for {num_participants} participant(s)...")
print("="*60)

# Run study for each participant
all_participant_data = []

for p_id in range(1, num_participants + 1):
    participant_data = run_complete_study(
        participant_id=p_id,
        movies_data=movies_df
    )
    
    # Save immediately after each participant
    save_participant_data(participant_data, filename="xai_study_results.csv")
    
    all_participant_data.append(participant_data)
    
    if p_id < num_participants:
        print(f"\n{'⏭️ '*30}")
        print(f"Moving to next participant ({p_id + 1}/{num_participants})...")
        print(f"{'⏭️ '*30}\n")
        time.sleep(2)

print("\n" + "🎊"*30)
print("="*60)
print("   ALL PARTICIPANTS COMPLETE!")
print("="*60)
print(f"✓ Data saved to: xai_study_results.csv")
print(f"✓ Total participants: {num_participants}")
print(f"✓ Total movie interactions: {num_participants * 12}")
print("🎊"*30 + "\n")
```
